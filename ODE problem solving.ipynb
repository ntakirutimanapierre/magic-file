```python
from scipy.integrate import odeint
import numpy as np
import matplotlib.pyplot as plt
```

# Question 1

```python
f=lambda x,y: ((1-x**2)**(-2)-4*x*y)/(1-x**2)
f1=lambda x,y: (y**4+x**3 *y)/x**4
x=np.linspace(1,10,100)
y1=odeint(f1,100,x)
y=odeint(f,0,x)
plt.figure(figsize=(15,4))
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,y1)
plt.show()
```

```python
f2=lambda x,y: (-y**2-4*x**2)/x*y
y2=odeint(f,0.000001,x)
plt.plot(x,y2)
plt.show()
```

```python
k=300
alpha=1
t=np.linspace(1,10,100)
T=lambda t,k:alpha*t*(1-t/k)

T1=odeint(T,10,t)
plt.plot(t,T1)
plt.show()
```

```python
beta=1
dxdt=lambda t,beta: 10*t*(1-t)
yy=odeint(dxdt,5,t)
plt.plot(t,yy)
plt.show()
```
